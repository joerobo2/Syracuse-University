```python
import os
import pandas as pd

# Dataset folder
dataset_path = "/root/.cache/kagglehub/datasets/bwandowando/ukraine-russian-crisis-twitter-dataset-1-2-m-rows/versions/510"
csv_files = [f for f in os.listdir(dataset_path) if f.endswith(".csv")]
csv_files = sorted(csv_files)

# Batch size (adjust as needed)
BATCH_SIZE = 10
all_batches = []

print(f"📂 Found {len(csv_files)} files. Processing in batches of {BATCH_SIZE}...")

for i in range(0, len(csv_files), BATCH_SIZE):
    batch_files = csv_files[i:i+BATCH_SIZE]
    batch_data = []

    for file_name in batch_files:
        file_path = os.path.join(dataset_path, file_name)
        print(f"📥 Loading: {file_name}")
        try:
            df = pd.read_csv(file_path, low_memory=False)
            df["source_file"] = file_name
            batch_data.append(df)
        except Exception as e:
            print(f"❌ Skipped {file_name}: {e}")

    # Concatenate this batch
    if batch_data:
        batch_df = pd.concat(batch_data, ignore_index=True)
        all_batches.append(batch_df)

        # Optional: clear memory
        del batch_data
        print(f"✅ Batch {i // BATCH_SIZE + 1} loaded. Current total rows: {sum([b.shape[0] for b in all_batches])}")

# Final combined DataFrame
df_all = pd.concat(all_batches, ignore_index=True)
print("\n✅ All batches combined.")
print("Final shape:", df_all.shape)
df_all.info()
```
